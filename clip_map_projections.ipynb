# Extracting Regions from Map Projections

The following examples demonstrate extracting regions from meshes defined on a 2D map projection.

```python
import pyvista as pv

from pv_utils import mesh_from_nc, project_mesh
```

```python
# Create Pyvista mesh from a netcdf file.
mesh = mesh_from_nc(fname='data/mesh_C12_synthetic.nc', data_type='synthetic')
mesh_30 = mesh.clip('y', invert=False, value=30)

p = pv.Plotter()
p.add_mesh(mesh, opacity=0.2,
           show_edges=True, edge_color='white', line_width=0.5,
           scalar_bar_args=dict(color='black', label_font_size=16,
                                n_labels=6, title_font_size=18))
p.add_mesh(mesh_30,
           show_edges=True, edge_color='white', line_width=0.5)

p.background_color = 'white'
p.camera_position = [(16.869657976834926, 5.235398397963124, 580.2717016329161),
                     (16.869657976834926, 5.235398397963124, 0.0),
                     (0.0, 1.0, 0.0)]
p.show()
```

```python
# Reproject the mesh from platecarree to a mollweide projection.
moll_mesh = project_mesh(mesh, 'moll')
moll_mesh_30 = project_mesh(mesh_30, 'moll')

# Plot the Mollweide projection.
p = pv.Plotter()
p.add_mesh(moll_mesh, opacity=0.2,
           show_edges=True, edge_color='white', line_width=0.5,
           scalar_bar_args=dict(color='black', label_font_size=16,
                                n_labels=6, title_font_size=18))
p.add_mesh(moll_mesh_30,
           show_edges=True, edge_color='white', line_width=0.5)
p.background_color = 'white'
p.camera_position = [(93959.85410932079, 328636.4707611543, 55954659.22149858),
                     (93959.85410932079, 328636.4707611543, 0.0),
                     (0.0, 1.0, 0.0)]
p.show()
```

```python
# Plot the Mollweide projection. ZOOMED
p = pv.Plotter()
p.add_mesh(moll_mesh, opacity=0.2,
           show_edges=True, edge_color='white', line_width=0.5,
           scalar_bar_args=dict(color='black', label_font_size=16,
                                n_labels=6, title_font_size=18))
p.add_mesh(moll_mesh_30,
           show_edges=True, edge_color='white', line_width=0.5)
p.background_color = 'white'
p.camera_position = [(93959.85410932079, 328636.4707611543, 25954659.22149858),
                     (93959.85410932079, 328636.4707611543, 0.0),
                     (0.0, 1.0, 0.0)]
p.show()
```
